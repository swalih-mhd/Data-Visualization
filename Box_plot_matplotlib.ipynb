```python
import matplotlib.pyplot as plt

class_A_scores=[55,60,62,65,67,70,72,75,50,85,90,95,98]
class_B_scores=[45,50,52,55,60,62,65,68,70,72,75,78,80]

plt.figure(figsize=(8,6))
plt.boxplot([class_A_scores,class_B_scores],labels=["class A","class B"])

plt.title("class score distribution")
plt.xlabel("classes")
plt.ylabel("scores")

plt.show()
```

```python
import matplotlib.pyplot as plt

data_1=[55,60,62,65,67,70,72,75,50,85,90,95,98,150]
data_2=[15,45,50,52,55,60,62,65,68,70,72,75,78,80]



fig,axs=plt.subplots(1,2,figsize=(12,6))

# boxplot for the first dataset
axs[0].boxplot(data_1)
axs[0].set_title("boxplot 1")
axs[0].set_xlabel("category")
axs[0].set_ylabel("values")

# boxplot for the second dataset
axs[1].boxplot(data_2)
axs[1].set_title("boxplot 2")
axs[1].set_xlabel("category")
axs[1].set_ylabel("values")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

data_1=[55,60,62,65,67,70,72,75,50,85,90,95,98,150]
data_2=[15,45,50,52,55,60,62,65,68,70,72,75,78,80]


plt.figure(figsize=(12,6))


# boxplot for the first dataset
plt.subplot(1,2,1)
plt.boxplot(data_1)
plt.title("boxplot 1")
plt.xlabel("category")
plt.ylabel("values")

# boxplot for the second dataset
plt.subplot(1,2,2)
plt.boxplot(data_2)
plt.title("boxplot 2")
plt.xlabel("category")
plt.ylabel("values")

plt.tight_layout()
plt.show()
```
