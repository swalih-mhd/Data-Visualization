#**violin plot**

```python
import matplotlib.pyplot as plt

data=[3,5,7,8,20,6,4,6]
plt.violinplot(data,showmeans=False,showmedians=True)

plt.title('simple violin plot')
plt.xlabel("category")
plt.ylabel('values')

plt.show()
```

# **violin plot with larger dataset**

```python
import matplotlib.pyplot as plt
import numpy as np

data=[np.random.normal(5,1,100),np.random.normal(7,0.5,100)]

plt.violinplot(data,showmedians=True)

plt.title('violin plot with larger dataset')
plt.xlabel("category")
plt.ylabel('values')
plt.show()
```

# **violin plot bodies properties**

```python
import matplotlib.pyplot as plt
import numpy as np

data=[np.random.normal(5,1,100),np.random.normal(7,0.5,100)]

violins=plt.violinplot(data,showmedians=True)

for x in violins['bodies']:
    x.set_facecolor('red')
    x.set_edgecolor('black')
    x.set_alpha(0.3)
    x.set_linewidth(2)

plt.title('violin plot with larger dataset')
plt.xlabel("category")
plt.ylabel('values')
plt.show()
```

# **violin plot with box plot**

```python
import matplotlib.pyplot as plt
import numpy as np

#np.random.seed(10)
data=[np.random.normal(5,1,100),np.random.normal(7,0.5,100)]

plt.violinplot(data,showmeans=True,showmedians=True)

plt.boxplot(data,widths=0.1,positions=[1,2])

plt.title('violin plot with box plot')
plt.xlabel("category")
plt.ylabel('values')
plt.show()
```

# **stem plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,10)
y=np.sin(x)

plt.stem(x,y,linefmt='-.',markerfmt='o',basefmt='k')

plt.title("stem plot example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0.1,2*np.pi,41)
y=np.exp(np.sin(x))

plt.stem(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2*np.pi,20)
y=np.sin(x)

plt.stem(x,y,basefmt=" ")

plt.title("stem plot example")
plt.xlabel("x-axis")
plt.ylabel("sin(x)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2*np.pi,15)
y=np.cos(x)

plt.stem(x,y,linefmt='r_',markerfmt='bo',basefmt='k',label="cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0.1, 2 * np.pi, 10)  # x values
y = np.sin(x)  # y values (sine function)

# Create stem plot
plt.stem(x, y, linefmt="--", markerfmt="o", basefmt="k")
plt.title("Simple Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis (sin(x))")
#plt.grid(True)

# Display plot
plt.show()
```
