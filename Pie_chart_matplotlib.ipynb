# **Matplotlib Pie Charts**

## **Creating Pie Charts**

With Pyplot, you can use the pie() function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])

plt.pie(y)
plt.show()
```

## **Labels**

Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

# **Start Angle**

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

# **Explode**

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0.2,0,0]

plt.pie(y, labels = mylabels,explode=myexplode)
plt.show()
```

# **Shadow**

Add a shadow to the pie chart by setting the shadows parameter to True:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

# **Colors**

You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

You can use Hexadecimal color values, any of the 140 supported color names, or one of these shortcuts:

'r' - Red

'g' - Green

'b' - Blue

'c' - Cyan

'm' - Magenta

'y' - Yellow

'k' - Black

'w' - White

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

# **Legend**

To add a list of explanation for each wedge, use the legend() function:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(loc="upper left")
plt.show()
```

# **Legend With Header**

To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```

## All in one code

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0.2,0,0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]




plt.pie(y, labels = mylabels,explode=myexplode,startangle = 90,shadow = True,colors=mycolors)
plt.legend(loc="upper left",title = "Four Fruits:")
plt.show()
```

```python
import matplotlib.pyplot as plt

outer_labels=['Fruits','Vegetables']
outer_sizes=[60,40]

inner_labels=['Apples','Bananas','Carrots','Broccoli']
inner_sizes=[30,30,20,20]

outer_colors=['skyblue','lightgreen']
inner_colors=['lightcoral','orange','yellowgreen','mediumseagreen']

fig,ax=plt.subplots()
ax.pie(outer_sizes,labels=outer_labels,colors=outer_colors,radius=1,wedgeprops=dict(width=0.3,edgecolor='w'))
ax.pie(inner_sizes,labels=inner_labels,colors=inner_colors,radius=0.7,wedgeprops=dict(width=0.3,edgecolor='w'))

plt.title('Nested Pie Chart Example')
plt.show()
```

# **tight_layout**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),label="sin(x)")
plt.plot(x,np.cos(x),label="cos(x)")
plt.legend(bbox_to_anchor=(1.05,1.0),loc="upper left")
plt.tight_layout()
plt.show()
```

# **frameon**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["sin wave","cos wave"],frameon=True)
plt.show()
```

## **overlaped stacked graph**

```python
x=[22,25,29,30,34,36,28,31,22,25,30,28]
y=[24,26,30,32,35,38,27,29,26,30,32,33]

bins=np.arange(20,40,5)#'overlaped stacked graph'
plt.hist(x,bins=bins,ec='k',alpha=0.7,label='Group A',stacked=True)
plt.hist(y,bins=bins,ec='k',alpha=0.7,label='Group B',stacked=True)
plt.title('overlaped stacked graph')
plt.xlabel('Age')
plt.ylabel=('Freaquency')
plt.legend()
plt.tight_layout()
plt.show()
```

# **overlayed histogram with overlapping axis labels**

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(50,10,100) # dataset 1: mean=50, SD=10, 100 samples
data2 = np.random.normal(60,15,100)# dtaset  2:mean= 60, SD=15, 100 samples

bins = np.linspace(20,100,20)# define bin range and size

plt.figure(figsize=(5,3))

# plot two histograms in the same figure
plt.hist(data1,bins=bins,color="blue",alpha=0.6,label="data1")
plt.hist(data2,bins=bins,color="orange",alpha=0.6,label="data2")


plt.xlabel("very long label for the x axis that might overlap with tick labels")
plt.ylabel("frequency of occurance (might overlap with bars)")
plt.title("overlayed histogram with overlapping axis labels")


# show the plot
plt.legend()

#plt.tight_layout
plt.show()
```

# **nested pie chart**

```python
import matplotlib.pyplot as plt

outer_labels=["Fruits","Vegetables"]
outer_sizes=[60,40]

inner_labels=["Apple",'Banana',"Carrot","Broccoli"]
inner_sizes=[30,30,20,20]

outer_colors=["skyblue","lightgreen"]
inner_colors=["lightcoral","orange","yellowgreen","mediumseagreen"]

fig,ax=plt.subplots()
ax.pie(outer_sizes,labels=outer_labels,colors=outer_colors,radius=1,wedgeprops=dict(width=0.3,edgecolor="w"))

ax.pie(inner_sizes,labels=inner_labels,colors=inner_colors,radius=0.7,wedgeprops=dict(width=0.3,edgecolor="w"))

plt.title("nested pie chart")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,13,18,16]

plt.plot(x,y,marker="o")


# add an annotation to the plot
plt.annotate("highest point",xy=(4,18),xytext=(3,20),arrowprops=dict(arrowstyle="->",color="red"))

plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("annotated line chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,13,17,16]

plt.plot(x,y,marker="o",color="b",ls="-",label="sample line")


# add an annotation to the plot
plt.annotate("highest point",xy=(4,17),xytext=(3.5,18),arrowprops=dict(arrowstyle="->",color="red"))

# annotate each data point wth its values
for i,j in zip(x,y):
  plt.text(i,j + 0.5,f"({i},{j})",ha="center",color="black")

plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("annotated line chart")
plt.legend()

plt.show()
```
