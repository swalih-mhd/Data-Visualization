## **diag_kind Parameter**

* Controls the type of plot used for the diagonal subplots (plots representing single-variable distributions).

**Possible values:**

auto (default): Automatically chooses the plot type based on kind.

hist: Creates histograms for the diagonal subplots.

kde: Creates kernel density estimation plots for single-variable distributions.

```python
import seaborn as sns

df=sns.load_dataset("tips")
sns.pairplot(df,hue="sex",markers=["o","s"],diag_kind="hist")
```

## **kind Parameter**

* Controls the type of plot used for off-diagonal (pairwise) relationships.

**Possible values:**

scatter (default): Creates scatter plots for pairwise relationships.

kde: Creates kernel density estimation plots to show the density of data points.

reg: Adds regression lines to the scatter plots.

```python
import seaborn as sns

df=sns.load_dataset("tips")
sns.pairplot(df,hue="sex",markers=["o","s"],kind="scatter")
```

# **Train-Test Split Visualization**


### **Concept of Train-Test Split**

The dataset is typically split into:

1.Training Set: Used to train the machine learning model (usually 70-80% of the data).

2.Test Set: Used to evaluate the model's performance on unseen data (usually 20-30%).

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# step 1: load the tips dataset
df=sns.load_dataset("tips")

# step 2: split the data into training and test sets
train_data, test_data=train_test_split(df, test_size=0.2,random_state=42)


sns.kdeplot(train_data["total_bill"],color="blue",label="Training Data")
sns.kdeplot(test_data["total_bill"],color="red",label="Test Data")


plt.xlabel("Total Bill")
plt.ylabel("Density")


plt.legend()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# step 1: load the tips dataset
df=sns.load_dataset("tips")

# step 2:prepare features (x) and target (y)
x=df.drop("tip",axis=1)
y=df["tip"]

# step 3: split the data into training and test sets
x_train,x_test,y_train,y_test,=train_test_split( x, y, test_size=0.2,random_state=42)

# step 4:visualize distributions for x_train and x_test
plt.figure(figsize=(10,5))

sns.kdeplot(x_train["total_bill"],color="blue",label="x_train (total_bill)")
sns.kdeplot(x_test["total_bill"],color="orange",label="x_test (total_bill)")

plt.title("comparison of total bill in x_train and x_test")
plt.legend()
plt.show()

sns.kdeplot(y_train,color="green",label="y_train")
sns.kdeplot(y_test,color="red",label="y_test")

plt.title("comparison of tip (y_train vs y_test)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
X=df[["total_bill","size"]]
Y=(df["sex"])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred,labels=[0,1])
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(cm,annot=True,fmt="d",xticklabels=le.classes_,yticklabels=le.classes_,cmap="viridis")
```

### **Multi-Class Classification**

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

df = sns.load_dataset("iris")

x = df[["sepal_length","sepal_width","petal_length","petal_width"]]
y = df["species"]

le=LabelEncoder()
y = le.fit_transform(y)


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

cm = confusion_matrix(y_test,y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt="d",xticklabels=le.classes_,yticklabels=le.classes_,cmap="Blues")

plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.title("Confusion Matrix")
plt.show()
```

# **ROC curve**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve,auc

# genarate synthetic data
x,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=42)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#train a Rabdom Forest Classifier
clf=RandomForestClassifier(random_state=42)
clf.fit(x_train,y_train)

# predict probabilities
y_probs=clf.predict_proba(x_test)[:,1]


fpr,tpr,thresholds=roc_curve(y_test,y_probs)
roc_auc=auc(fpr,tpr)

# plot the ROC curve
plt.figure(figsize=(8,6))
sns.lineplot(x=fpr,y=tpr,label=f"ROC curve (AUC={roc_auc:.2f})",color="blue")

plt.plot([0,1],[0,1],linestyle="--",color="gray",label="Random Guess") #diagnal line

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve,auc,roc_auc_score

# step 1: load the iris dataset and prepare binary labels
df=load_iris()
x=df.data
y=(df.target==2).astype("int")

# step 2:split the datasets into  training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# step 3: train a LogisticRegression model
model=LogisticRegression(random_state=42)
model.fit(x_train,y_train)

# step 4:

y_probs=model.predict_proba(x_test)[:,1]

# step 5:compute the ROC curve and AUC
fpr,tpr,thresholds=roc_curve(y_test,y_probs)
auc_score=roc_auc_score(y_test,y_probs)

#step 6:plot the ROC curve using seaborn styling
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f"ROC curve (AUC={auc_score:.2f})",color="blue",lw=2)
plt.plot([0,1],[0,1],linestyle="--",color="gray",label="Random Classifier")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```

# **Precision-Recall Curve**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.datasets import load_iris

# step 1: load the iris dataset and prepare binary labels
iris = load_iris()
x=iris.data
y=(iris.target ==2).astype('int')

# step 2: plit the l
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

# step 3: train the model
model=LogisticRegression(random_state=42)


model.fit(x_train,y_train)

# step4 :
y_probs=model.predict_proba(x_test)[:,1]

#step5: cumpute the precision-recall curve and average precision score
precision,recall,threshould=precision_recall_curve(y_test,y_probs)
average_precision = average_precision_score(y_test,y_probs)



# step 6: plot the ROC Curve using seaborn stylling
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,6))
plt.plot(recall,precision,label=f'precision recall Curve (ap;={average_precision:.2f})',color="blue",lw=2)
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('precision recall Curve')
plt.legend(loc='lower left')
plt.show()
```

# **Residual Plots**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

tips=sns.load_dataset('tips')

X=tips[['total_bill']]
y=tips['tip']

# train the linearregression model
model=LinearRegression()
model.fit(X,y)

# predict tips using the model
y_pred=model.predict(X)

#calculate residuals
residuals=y-y_pred

#create a residentional plot
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,6))
sns.residplot(x=y_pred,y=residuals,color="blue",lowess=True,line_kws={'color':'red'})
plt.axhline(0, linestyle='--',color='gray',linewidth=1) # Reference line at 0
plt.xlabel("Predicted Tips")
plt.ylabel("Residuals")
plt.title("Residual Plot: tips dataset (Linear Regression)")
plt.show()
```
