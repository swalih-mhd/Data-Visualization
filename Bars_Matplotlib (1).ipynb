# **Matplotlib Bars**

## **Creating Bars**
With Pyplot, you can use the bar() function to draw bar graphs:


```python
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

The bar() function takes arguments that describes the layout of the bars.

The categories and their values represented by the first and second argument as arrays.





```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
plt.xlabel("fruits",loc="left")
plt.ylabel("price")
plt.title("vertical bar graph")
plt.show()
```

# **Horizontal Bars**

If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(2, 2)

# Add titles and labels
for i, ax in enumerate(axs.flat):
    ax.set_title(f"Plot {i+1}")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.xlabel("price")
plt.ylabel("fruits")
plt.title("horizontal bar graph")
plt.show()
```

# **Bar Color**

The bar() and barh() take the keyword argument color to set the color of the bars:

## Color Hex

Or you can use Hexadecimal color values:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

# **Bar Width**

The bar() takes the keyword argument width to set the width of the bars:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y,width=0.5)
plt.show()
```

## **Simple Stacked Bar Plot**

```python
import matplotlib.pyplot as plt

x=["A","B","C","D"]
y1=[10,20,10,30]
y2=[20,25,15,25]

plt.bar(x,y1,color="red")
plt.bar(x,y2,bottom=y1,color="b")
plt.show()
```

# **Bar Height**

The barh() takes the keyword argument height to set the height of the bars:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([3,6,9,4,2,5])
y=np.arange(6)
plt.bar(y,x,color="yellow")
plt.bar(y,-x,color="g")
plt.title("back to back bar")
plt.show()
```

## **multiple bars**

```python

import matplotlib.pyplot as plt
import numpy as np

# set width of bar
barwidth=0.25
fig=plt.subplots(figsize=(10,4))

# set hight of bar on x axis
IT=[12,30,1,8,22]
ECE=[28,6,16,5,10]
CSE=[29,3,24,25,17]

# set position of bar on x axis
br1=np.arange(len(IT)) # 0,1,2,3,4
br2=[x+barwidth for x in br1] # 0.25,1.25,2.25,3.25,4.25
br3=[x+barwidth for x in br2]

plt.bar(br1,IT,color="r",width=barwidth,edgecolor="grey",label="IT")
plt.bar(br2,ECE,color="g",width=barwidth,edgecolor="grey",label="ECE")
plt.bar(br3,CSE,color="b",width=barwidth,edgecolor="grey",label="CSE")


# adding xticks
plt.xlabel("branch",fontweight="bold",fontsize=10)
plt.ylabel("studnts passed",fontweight="bold",fontsize=10)
plt.xticks([r + barwidth for r in range(len(IT))],["2015","2016","2017","2018","2019"])
# syntax: xticks(ticks,labels)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N=5

boys=(20,35,30,35,27)
girls=(25,32,34,20,25)
boystd=(2,3,4,1,2)
girlstd=(3,5,2,3,3)
ind=np.arange(N)
width=0.5

fig=plt.subplots(figsize=(10,7))
p1=plt.bar(ind,boys,width,yerr=boystd)
p2=plt.bar(ind,girls,width,bottom=boys,yerr=girlstd)

plt.ylabel("Contribution")
plt.title("Contribution by the teams")
plt.xticks(ind,("T1","T2","T3","T4","T5"))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),("boys","girls"))

plt.show()
```
